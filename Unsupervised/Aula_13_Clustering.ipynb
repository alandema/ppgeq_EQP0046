```python
import matplotlib.pyplot as plt
#from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

### Exemplo 1

```python
features, true_labels = make_blobs(n_samples=200, centers=3, cluster_std=2.75, random_state=42)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
kmeans = KMeans(init="random", n_clusters=3, n_init=10, max_iter=300, random_state=42)
```

```python
kmeans.fit(scaled_features)
```

```python
clust = kmeans.cluster_centers_
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(scaled_features[:,0],scaled_features[:,1],c = kmeans.labels_)
scatter = ax.scatter(clust[:,0],clust[:,1],marker='x',s=200,c='black')
```

### Exemplo 2

```python
from sklearn.cluster import DBSCAN
```

```python
clustering = DBSCAN(eps=0.5, min_samples=5).fit(scaled_features)
```

```python
_, ax = plt.subplots()
scatter = ax.scatter(scaled_features[:,0],scaled_features[:,1],c = clustering.labels_)
```
